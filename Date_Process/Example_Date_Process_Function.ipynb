```python
import numpy as np
#Use skimage package to deal image
from skimage import io
from skimage.io import imread
import matplotlib.pyplot as plt

from data_process import label2colar,result2image,resize_image_to_256,resize_result_to_original
```

```python
#Load the exmaple dats
result_array = np.load('/content/orange_3_doors_0001.npy')
original_image_noseg = imread('/content/3doors_noseg.png')
original_image_seg = imread('/content/3doors_seg.png')
```

```python
# Show result_array by image
fig, ax = plt.subplots(1,3, figsize=(15,10))
ax[0].imshow(result_array[:,:,0:3])
ax[0].set_title('Image')

#use the label2colar to show the segmentation result
ax[1].imshow(label2colar(result_array[:,:,3]))
ax[1].set_title('Segmentation')

ax[2].imshow(result2image(result_array))
ax[2].set_title('Image with Segmentation')

plt.show()
```

```python
#Resize the original image to the particular size so that we can put it in the net
#Shape of input of the U_net I built is 256*256*3

#Change the image shape to 256*256*3
resized_image = resize_image_to_256(original_image_noseg)
print("The original image shape is", original_image_noseg.shape)
print("The resized image shape is", resized_image.shape)

#Show original image and resize image
fig, ax = plt.subplots(1,2, figsize=(15,10))
ax[0].imshow(original_image_noseg)
ax[0].set_title('Original_Image')
ax[1].imshow(resized_image)
ax[1].set_title('Resized_image')
plt.show()
```

```python
#example of resize_result_to_original function
fig, ax = plt.subplots(1,2, figsize=(15,10))
ax[0].imshow(original_image_seg)
ax[0].set_title('Given')
#Add the segmentation result to the original image
ax[1].imshow(resize_result_to_original(original_image_noseg,result_array))
ax[1].set_title('Resize_result')
plt.show()
```
